```python
# (0) function to check-print all functions in the notebook
def print_total_functions():
  # Get all the global variables in the current namespace
  global_vars = globals()

  # Filter out the functions
  functions = [name for name, obj in global_vars.items() if inspect.isfunction(obj)]
  # Print the total number of functions
  print(f"Total functions = {len(functions)}\n")

  # Enumerate and print the list of function names
  print("Defined functions:")
  for index, func in enumerate(functions):
      print(f"({index}, {func})")

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (1) define a function to load the data
def load_housing_data():
  tarball_path = Path("datasets/housing.tgz")
  if not tarball_path.is_file():
    Path("datasets").mkdir(parents=True, exist_ok=True)
    url = "https://github.com/ageron/data/raw/main/housing.tgz"
    urllib.request.urlretrieve(url, tarball_path)
  with tarfile.open(tarball_path) as housing_tarball:
    housing_tarball.extractall(path="datasets")
  return pd.read_csv(Path("datasets/housing/housing.csv"))

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (2) define the default font sizes
plt.rc('font',size=14)
plt.rc('axes',labelsize=14, titlesize=14)
plt.rc('legend',fontsize=14)
plt.rc('xtick',labelsize=10)
plt.rc('ytick',labelsize=10)

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (3) enable dark mode for plots
plt.style.use('dark_background')

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (4) set random seed value
np.random.seed(42)
random_state_val = 42

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (5) set test_size value
test_size_val = .2
# +--------------------------------------------------------------------------------------------------------------------------------------+

# (6) function to check incomed category proportions between random sampling test set, stratified test set and overall test set
def income_cat_proportions(data):
  return data["income_cat"].value_counts() / len(data)

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (7) function to apply the log_transform to a dataset (for prescaling skewness treatment)
log_transformer = FunctionTransformer(np.log, inverse_func=np.exp)

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (8) example custom rbf_kernel function: to measure the geographic similarity between each district and San Francisco

# san francisco coordinates
sf_coords = 37.7749,-122

# function writing
sf_transformer = FunctionTransformer(rbf_kernel,
                                     kw_args=dict(
                                         Y=[sf_coords], gamma=.1
                                                  ))

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (9) custom transformer that acts much like the StandardScaler
class StandardScalerClone(BaseEstimator, TransformerMixin):
  def __init__(self, with_mean=True): # no *args or **kwargs
    self.with_mean = with_mean

  def fit(self, X, y=None):
    X = check_array(X) # checks that X is an array with finite float values
    self.mean = X.mean(axis=0)
    self.scale_ = X.std(axis=0)
    self.n_features_in = X.shape[1] # every estimator stores this in fit()
    return self

  def transform(self, X):
    check_is_fitted(self) # looks for learned features (with trailing _)
    X = check_array(X)
    assert self.n_features_in == X.shape[1]
    if self.with_mean:
      X = X - self.mean_
    return X / self.scale_

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (10) custom transformer for cluster similarity
class ClusterSimilarity(BaseEstimator, TransformerMixin):
    def __init__(self, n_clusters=10, gamma=1.0, random_state=None):
        self.n_clusters = n_clusters
        self.gamma = gamma
        self.random_state = random_state

    def fit(self, X, y=None, sample_weight=None):
        self.kmeans_ = KMeans(self.n_clusters, n_init=10,
                              random_state=self.random_state)
        self.kmeans_.fit(X, sample_weight=sample_weight)
        return self

    def transform(self, X):
        return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)

    def get_feature_names_out(self, names=None):
        return [f"Cluster {i} similarity" for i in range(self.n_clusters)]

# +--------------------------------------------------------------------------------------------------------------------------------------+

# (11) monkey patch for get_feature_names_out() due to version
_monkey_patched_already = False

def monkey_patch_get_signature_names_out():
    """
    Monkey patch some classes which did not handle get_features_names_out()
    correctly in Scikit-Learn 1.0.x.
    """
    global _monkey_patched_already

    if _monkey_patched_already:
        # Already patched, do nothing and skip re-patching
        return

    default_get_feature_names_out = StandardScaler.get_feature_names_out

    if not hasattr(SimpleImputer, "get_feature_names_out"):
        print("Monkey-patching SimpleImputer.get_feature_names_out()")
        SimpleImputer.get_feature_names_out = default_get_feature_names_out

    if not hasattr(FunctionTransformer, "get_feature_names_out"):
        print("Monkey-patching FunctionTransformer.get_feature_names_out()")

        orig_init = FunctionTransformer.__init__
        orig_sig = signature(orig_init)

        # Only patch if 'feature_names_out' is not already in the signature
        if 'feature_names_out' not in orig_sig.parameters:
            def __init__(self, *args, feature_names_out=None, **kwargs):
                bound = orig_sig.bind(self, *args, **kwargs)
                bound.apply_defaults()
                orig_init(self, *args, **kwargs)
                self.feature_names_out = feature_names_out

            __init__.__signature__ = Signature(
                list(orig_sig.parameters.values()) + [
                    Parameter("feature_names_out", Parameter.KEYWORD_ONLY)
                ]
            )

            def get_feature_names_out(self, names=None):
                if callable(self.feature_names_out):
                    return self.feature_names_out(self, names)
                assert self.feature_names_out == "one-to-one"
                return default_get_feature_names_out(self, names)

            FunctionTransformer.__init__ = __init__
            FunctionTransformer.get_feature_names_out = get_feature_names_out

    _monkey_patched_already = True
```

### **Implement *Stratified Sampling* with an Income Category**
The overall population will notably display ratios in its data that need to be reflected in the sample.

The sample population divides into several homogenous groups called *strata*, and the right number of instances are sampled from each stratum, guaranteeing that the test set is representative of the overall population.  

```python
# create an income category to ensure that the test set is represntative of various categories of income
housing["income_cat"] = pd.cut(housing["median_income"],
                              bins=[0.,1.5,3.0,4.5,6.,np.inf],
                              labels=[1,2,3,4,5])

housing["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid = True, color='red')
plt.grid(alpha=.5)
plt.xlabel("Income category")
plt.ylabel("Number of districts")
plt.title("Income categories: stratified sampling\n")
plt.show()
```

```python
# stratified split() method
splitter = StratifiedShuffleSplit(n_splits=10, test_size=test_size_val, random_state=random_state_val)
strat_splits = []
for train_index, test_index in splitter.split(housing, housing["income_cat"]):
  strat_train_set_n = housing.iloc[train_index]
  strat_test_set_n = housing.iloc[test_index]
  strat_splits.append([strat_train_set_n, strat_test_set_n])
```

```python
strat_train_set, strat_test_set = train_test_split(
    housing, test_size = test_size_val, stratify=housing["income_cat"],
    random_state=random_state_val
)
```

```python
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

```python
# sampling bias comparison of stratified versus purely random sampling
train_set, test_set = train_test_split(housing, test_size=test_size_val, random_state=random_state_val)

compare_props = pd.DataFrame({
    "Overall %": income_cat_proportions(housing),
    "Stratified %": income_cat_proportions(strat_test_set),
    "Random %": income_cat_proportions(test_set),
}).sort_index()
compare_props.index.name = "Income Category"
compare_props["Strat. Error %"] = (compare_props["Stratified %"] /
                                   compare_props["Overall %"] - 1)
compare_props["Rand. Error %"] = (compare_props["Random %"] /
                                  compare_props["Overall %"] - 1)
(compare_props * 100).round(2)
```

```python
# drop the income category, as it won't be used again
for set_ in (strat_train_set, strat_test_set):
  set_.drop("income_cat", axis=1, inplace=True)
```

* **KNNImputer**: replaces each missing value with the mean of the K-nearest neighbours' values for that feature. The distance is based on all available features in the dataset.

* **IterativeImputer**: trains a regression model per feature to predict the missing values based on all the other available features. It then trains the model again on the updated training data, and repeats the process several times over, improving the models and the value predictions iteratively.

*Supporting non-numerical values:*

* **imputer = SimpleImputer(strategy="most_frequent")**
* **imputer = SimpleImputer(strategy="constant, fill_value= n")**

```python
# generate a cat_pipeline and use a ColumnTransformer() instance to have a unified pipeline in a single variable
cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore")
)

preprocessing = make_column_transformer(
    (num_pipeline, make_column_selector(dtype_include=np.number)),
    (cat_pipeline, make_column_selector(dtype_include=object)),
)
```

```python
# update the pipeline to include the subsystems: ratio_pipeline, log_pipeline, cluster_simil, cat_pipeline
def column_ratio(X):
  return X[:, [0]] / X[:, [1]]

def ratio_name(function_transformer, feature_names_in):
  return ["ratio"] # feature names out

def ratio_pipeline():
  return make_pipeline(
      SimpleImputer(strategy="median"),
      FunctionTransformer(column_ratio, feature_names_out=ratio_name),
      StandardScaler()
  )

# logarithm transformations
log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(np.log, feature_names_out="one-to-one"),
    StandardScaler()
)
# soft coding
gamma_val=1.0
clusters_val=10

# clusters feature
cluster_simil = ClusterSimilarity(n_clusters=clusters_val, gamma=gamma_val, random_state=random_state_val)

# numerical pipeline
default_num_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)

# preprocessing pipeline
preprocessing = ColumnTransformer([
    ("bedrooms", ratio_pipeline(),["total_bedrooms", "total_rooms"]),
    ("rooms_per_house", ratio_pipeline(),["total_rooms", "households"]),
    ("people_per_house", ratio_pipeline(),["population","households"]),
    ("log", log_pipeline, [
        "total_bedrooms",
        "total_rooms",
        "population",
        "households",
        "median_income",
    ]),
    ("geo", cluster_simil, ["latitude","longitude"]),
    ("cat", cat_pipeline, make_column_selector(dtype_include=object)),
],
remainder=default_num_pipeline) # one column remaining: housing_median_age
```
